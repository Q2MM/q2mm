```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
from_brock = pd.read_csv('data_compare.csv')
from_brock
```

```python
import seaborn as sns
```

```python
from_brock['combo'] = from_brock['lig']+from_brock['sub']
from_brock
```

```python
sns.set_style('white')
sns.set_context('notebook')
sns.set_theme(rc={'figure.figsize':(11.7, 8.27)})
ee_diff_plot = sns.barplot(data=from_brock, x='combo', y='ee_diff', hue='d')
sns.move_legend(ee_diff_plot, "lower center", bbox_to_anchor=(.5, 1), ncol=2, title=None, frameon=False)
ee_diff_plot.add_line()
```

```python
ddee_diff_plot = sns.barplot(data=from_brock, x='combo', y='dde_diff', hue='d')
sns.move_legend(ddee_diff_plot, "lower center", bbox_to_anchor=(.5, 1), ncol=2, title=None, frameon=False)
```

```python
melted = pd.melt(from_brock, id_vars=['combo', 'd'], value_vars=['ee_diff', 'dde_diff', 'ee', 'dde', 'Ocalc_dde', 'Ocalc_ee'])
melted
sns.barplot(data=melted, x='variable', y='value', hue='d', ci=None)
```

```python
filtered = from_brock[from_brock['combo'] != 'B3']
filtered = filtered[filtered['combo'] != 'E3']
filtered = filtered[filtered['combo'] != 'G3Z']
print(filtered['combo'].values)
melted_2 = pd.melt(filtered, id_vars=['combo', 'd'], value_vars=['ee_diff', 'dde_diff', 'ee', 'dde', 'Ocalc_dde', 'Ocalc_ee'])
```

```python
sns.barplot(data=melted_2, x='variable', y='value', hue='d', ci=None)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,10))
brock_best = from_brock[from_brock['d']=='RUN-BEST_BROCK_FF2']
donoghue = from_brock[from_brock['d']=='RUN-rh_hyd_enamide_donoghue_final']
q_fuerza = from_brock[from_brock['d']=='RUN-rh_hyd_enamide_q_seminario']
ho_local = from_brock[from_brock['d']=='RUN-rh_qf_localized_gbest_x_2_0.878']


slope, intercept, r2, pv, se = stats.linregress(brock_best['exp_ee'], brock_best['ee'])
sns.regplot(brock_best, x='exp_ee', y='ee', ax=ax, label='brock_best', line_kws={'label':'$y=%3.7s*x+%3.7s   r2:%3.7s$'%(slope, intercept, r2)}, ci=None)
slope, intercept, r2, pv, se = stats.linregress(donoghue['exp_ee'], donoghue['ee'])
sns.regplot(donoghue, x='exp_ee', y='ee', ax=ax, label='donoghue', line_kws={'label':'$y=%3.7s*x+%3.7s   r2:%3.7s$'%(slope, intercept, r2)}, ci=None)
slope, intercept, r2, pv, se = stats.linregress(q_fuerza['exp_ee'], q_fuerza['ee'])
q_f_plot = sns.regplot(q_fuerza, x='exp_ee', y='ee', ax=ax, label='q_fuerza', line_kws={'label':'$y=%3.7s*x+%3.7s   r2:%3.7s$'%(slope, intercept, r2)}, ci=None)
slope, intercept, r2, pv, se = stats.linregress(ho_local['exp_ee'], ho_local['ee'])
sns.regplot(ho_local, x='exp_ee', y='ee', ax=ax, label='loc_best_0.878', line_kws={'label':'$y=%3.7s*x+%3.7s   r2:%3.7s$'%(slope, intercept, r2)}, ci=None)
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x=q_f_plot.get_lines()[0].get_xdata(),y=q_f_plot.get_lines()[0].get_ydata())
ax.legend()
sns.move_legend(ax, loc='best')
ax.set_ylim(bottom=-110, top=110)
ax.set_xlim(left=-110, right=110)
```
