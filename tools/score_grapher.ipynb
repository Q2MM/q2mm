```python
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
import copy
seaborn.set_theme(style="whitegrid", palette="pastel")
```

```python
col_names = ["Label", "Weight",   "Reference",   "Calculated",      "Score", "FF"]
q_sem_eig = pd.read_csv("../rh-seminario/mm3_results/q_seminario_mjeigz_score.txt", names=col_names)
q_sem_eig.assign(FF='q_sem')
old_initls_eig = pd.read_csv("../rh-seminario/mm3_results/old_initials_mjeigz_score.txt", names=col_names)
old_initls_eig.assign(FF='old_init')
donoghue_eig = pd.read_csv("../rh-seminario/mm3_results/donoghue_mjeigz_score.txt", names=col_names)
donoghue_eig.assign(FF='donoghue')
```

```python
print(q_sem_eig.head())
merged = copy.deepcopy(q_sem_eig)
merged.append(old_initls_eig)
merged.append(donoghue_eig)
print(merged.head())
```

```python
ax = seaborn.regplot(data=q_sem_eig, x='Reference', y='Calculated', fit_reg=False)
seaborn.regplot(data=donoghue_eig, x='Reference', y='Calculated', color='green', ax=ax, fit_reg=False)
seaborn.regplot(data=old_initls_eig, x='Reference', y='Calculated', color='red', ax=ax, fit_reg=False)
ax.set_ylim(top=5000)
ax.set_xlim(xmin=-2000, xmax=5000)
```
